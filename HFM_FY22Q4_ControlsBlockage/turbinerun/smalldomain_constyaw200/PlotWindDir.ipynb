### Plot at 1,000 seconds in

```python
x,y,z,s1,s2,vx = ppsample.getPlaneSampleAtTime(wdir.sample_ncdat, 'p_hub', 'velocityx', 40, 1)
x,y,z,s1,s2,vy = ppsample.getPlaneSampleAtTime(wdir.sample_ncdat, 'p_hub', 'velocityy', 40, 1)
```

```python
vh = np.sqrt(vx**2 + vy**2)

clevels=np.linspace(0,11,81)
plt.figure(figsize=(6,5), dpi=125)
plt.contourf(x, y, vh, clevels, cmap='coolwarm')
plt.colorbar()
plt.axis('scaled')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('horizontal U, t=%0.1f, z=150m'%(40*25+20000))
```
